Al ver que hay distintos tipos de datos, pasamos a convertirlo en uno solo, en este caso un dict. Para ello se usa ast.literal_eval(x) aprovechando que en nuestras columnas a desanidar tienen formato de diccionarios o listas de diccionarios, como es posible encontrar con valores NaN, para evitar que generen un error a futuro se cambian por diccionario vacios.

```python
def convertir_a_diccionario(df,columna): # se crea una función que tome como parametro un dataframe y una columna dentro del mismo
    df[columna] = df[columna].apply( #se aplica la función apply para que la condición se aplique a toda la columna
    lambda x: ast.literal_eval(x) if isinstance(x, str) else {} # la condición es que si el dato es str, aplica ast.literal_eval, 
    #volviendolo un diccionario en este caso, de lo contrario devuelve un diccionario vacio.
)
```

```python
convertir_a_diccionario(data,'belongs_to_collection') #usamos la función
```

```python
print(data['belongs_to_collection'].apply(type)) #Verificamos que la función de los resultados esperados
```

```python
# Creamos una función con el metodo apply para desanidar la columna 'belongs_to_collection' ya que logramos que todos sus datos sean dict.
def desanidar_dict(df,columna):
    desanidar_data= df[columna].apply(pd.Series)
    return desanidar_data
```

##### 1.2- DESANIDAR 'genres' 

```python
# Revisar que tipo de dato tenemos. 

print(data['genres'].head(3).apply(type)) # Revisamos que haya un mismo tipo de dato en común.
```

```python
#creamos una función que tome como parametro cadenas de texto
def desanidar_list_dic(cadena_de_diccionarios):
    ''' 
    Esta función permite tomar un texto y devolver los valores de la clave 'name'
    '''
    valores_name = []  # se crea una lista donde se guardarán los valores de la clave 'name'
    
    if pd.isna(cadena_de_diccionarios) or cadena_de_diccionarios == '': # se pone un filtro y se usa isna para saber si hay un valor nan o un string vacio
        return ['No hay datos']  # si el valo resulta ser nan o vacio se devuelve lo siguiente
    try:
        for palabra in ast.literal_eval(cadena_de_diccionarios): # Se crea un for para recorrer toda la cadena que con ayuda de ast.literal_eval se convertirá en una lista de diccionarios
            valores_name.append(palabra['name'])  # agregar los valores de 'name'
    except (ValueError, SyntaxError) as e:
        valores_name.append('Error en la conversión')  #utilizamos el bloque except para que si se da algún error de conversión como ValueError, no detenga el programa y de un aviso 

    return valores_name #retornamos la lista donde guardamos los valores de name. 
```

```python
data['genres']=data['genres'].apply(desanidar_list_dic) #aplicamos la función y reemplazamos los valores original dentro de la columna original
```

##### 1.6 Desanidar 'crew'

```python
print(data['crew'].apply(type)) #vemos su tipo de dato
```

```python
data['crew'].iloc[0] # Para este caso solo necesitamos al director, por lo que vemos como estan estructuradas las filas, tomando de ejemplo la primera.  
```

```python
# Para este caso creamos un función nueva. 
def desanidar_list_director(cadena_crew):
    ''' 
    Esta función permite tomar una cadena de texto, la recorre y devuelve el nombre del director
    '''
    directores = []  # Creamos una lista para guardar a los directores
    try:
        lista_crew = ast.literal_eval(cadena_crew) #usamos ast.literal_eval aprovechando la estructura de los datos para volverla una lista de dic.
        for miembro in lista_crew: # se recorre la lista de diccionarios.
            if miembro.get('job') == 'Director':  # se aplica el primer filtro con .get para que se encuentre la primera condición que es que su trabajp
                # sea de director
                directores.append(miembro['name'])  # Una vez se cumple la condición se busca la clave name dentro de ese diccionario y se
                # agrega su valor a la lista directores que es la que se devolvera al final.
    except (ValueError, SyntaxError) as e:
        print(f'Error en la conversión de la cadena: {e}') # Se pone lo siguiente para que si hay un error de conversión no detenga el programa.
    
    return directores  
```

```python
data['crew']= data['crew'].apply(desanidar_list_director) #asignamos los valores desanidados a la columna original
```

```python
data.loc[:, 'popularity'] = pd.to_numeric(data['popularity'], errors='coerce') #pasamos popularity a un solo tipo de dato númerico
```

```python
peliculas_viejas_no_populares = [] #Se crea una lista vacia para almacer los indices que arroje el for

for indice, columna in data.iterrows(): #se hace un for para recorrer los valores de las filas
    if columna['release_year'] < 2000 and columna['popularity'] < 4.0: # se establece las condiciones de que si la pelicula fue lanzada antes del 2000 y su popularidad
        peliculas_viejas_no_populares.append(indice) # es menor a 4.0, se agrega a la lista de peliculas_viejas_no populares. 
```

```python
print(len(peliculas_viejas_no_populares)) #vemos cuantas peliculas cumples la condición
```

```python
data= data.drop(index=peliculas_viejas_no_populares, axis= 1) # utilizamos drop para borrar las filas, se usa index por que estamos con ellos y ya los tenemos 
# identificados en la lista peliculas_viejas_no_populares 
```
